## 13장 모델 성능 검증하기
- 측정된 정확도를 어떻게 신뢰할 수 있는지, 예측 결과의 신뢰도를 높이는 방법을 알아보자

```python
import pandas as pd
```

```python
# 이진분류
# 입력feature 60개, target 1개
# header가 없는 경우 header=None
# data와 .ipynb는 같은 폴더에 있어야함
df = pd.read_csv('data/sonar3.csv', header=None)
df.head()
```

```python
# 광물111개, 일반암석97개
df[60].value_counts()
```

```python
# X: 입력feature, y: 광물의 종류 저장
X = df.iloc[:, :60]  
y = df.iloc[:, 60]
```

```python
from tensorflow.keras import models, layers
```

```python
model = models.Sequential()
model.add(layers.Dense(24, input_dim=60, activation='relu'))
model.add(layers.Dense(10, activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
# 손실함수와 optimizer 장착, 척도
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X, y, epochs=200, batch_size=10)
```

- accuracy 결과 100% → 이는 모델의 일반화가 안되어 있음. 훈련데이터에 완전 맞춰져 있음.   
  **=> 과적합(overfitting)**   
  **=> 훈련셋, 테스트셋 나눠야한다!!**
