- EarlyStopping   
   * 학습이 진행되어도 테스트셋 오차가 줄어들지 않으면 학습을 자동으로 멈추게 하는 함수
   * 검증셋의 오차가 커지기 전에 학습을 자동으로 중단시키고, 그때의 모델을 저장하는 방법

```python
from tensorflow.keras import models, layers
from sklearn.model_selection import train_test_split
import pandas as pd
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
import os
```

```python
df = pd.read_csv('data/wine.csv', header=None)
X = df.iloc[:, :12]
y = df.iloc[:, 12]
X_train, X_test, y_train, y_test = \
train_test_split(X, y, test_size=0.2, shuffle=True)
```

```python
model = models.Sequential()
model.add(layers.Dense(30, input_dim=12, activation='relu'))
model.add(layers.Dense(12, activation='relu'))
model.add(layers.Dense(8, activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# val_loss 20번 이상 좋아지지 않으면(감소하지 않으면) 강제로 학습을 중지시켜라
early_stopping = EarlyStopping(monitor='val_loss', patience=20)
model_path = './model/best_model.hdf5'
checkpoint = ModelCheckpoint(filepath=model_path, monitor='val_loss', verbose=0,
                            save_best_only=True)
# 결과가 좋아졌을 때만(전 회차 val_loss보다 적어졌을 때) overwrite해라
```

```python
history = model.fit(X_train, y_train, epochs=2000, batch_size=500, verbose=1,
                    validation_split=0.25, callbacks=[early_stopping, checkpoint])
```

```python
score = model.evaluate(X_test, y_test)
print('Test Accuracy: ', score[1])
```
