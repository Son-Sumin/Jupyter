```python
import pandas as pd
```

```python
df = pd.read_csv('data/sonar3.csv', header=None)   # header가 없는 경우   # data와 .ipynb는 같은 폴더에 있어야함
df.head()
```

```python
from tensorflow.keras import models, layers
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:, :60]
y = df.iloc[:, 60]
```

```python
X_train, X_test, y_train,y_test = \
train_test_split(X, y, test_size=0.3, shuffle=True)
```

```python
model = models.Sequential()
model.add(layers.Dense(24, input_dim=60, activation='relu'))
model.add(layers.Dense(10, activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))
```

```python
# 손실함수와 optimizer 장착, 척도
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=200, batch_size=10)
```

```python
score = model.evaluate(X_test, y_test)  # tensorflow.keras에서 제공하는 evaluate 
print('Test Accuracy: ', score[1])  # score[1] : 정확도
print('Test Loss: ', score[0])  # score[0] : 손실률
```
