```python
from bs4 import BeautifulSoup
```

```python
html = """<html><body><div><span>
        <a href=http://www.naver.com>naver</a>
        <a href=https://www.google.com>google</a>
        <a href=http://www.daum.net/>daum</a>
    </span></div></body></html>"""
```

```python
# BeautifulSoup(긁어온 html 소스, 파서)
soup = BeautifulSoup(html, 'lxml')
```

```python
print(soup)
```

```python
# 구조적으로 출력 .prettify()
print(soup.prettify())
```

```python
# 특정 태그 출력 방법1 : .find()
print(soup.find('a'))
```

```python
# 특정 태그 내용만 출력 : .find().get_text()
print(soup.find('a').get_text())
```

```python
# 특정 태그 모두 출력 (결과 : 리스트 형태) :.find_all()
print(soup.find_all('a'))
```

```python
# 특정 태그 모두의 모든 내용만 출력  -> 프로그램적으로 해결하기
sites = soup.find_all('a')
for site in sites:
    print(site.get_text())
```

```python
html2 = """
<html>
 <head>
  <title>작품과 작가 모음</title>
 </head>
 <body>
  <h1>책 정보</h1>
  <p id="book_title">토지</p>
  <p id="author">박경리</p>

  <p id="book_title">태백산맥</p>
  <p id="author">조정래</p>

  <p id="book_title">감옥으로부터의 사색</p>
  <p id="author">신영복</p>
 </body>
</html>
"""
```

```python
soup2 = BeautifulSoup(html2, 'lxml')
```

```python
# 특정 태그 출력 방법2
print(soup2.title)
print(soup2.body)
print(soup2.body.h1)
```

```python
# 특정 태그의 특정 아이디 맨 첫 번째 것만 출력
print(soup2.find('p', {'id': 'book_title'}))
```

```python
# 특정 태그의 특정 아이디 모두 출력
print(soup2.find_all('p', {'id': 'book_title'}))
```

```python
# 특정 태그의 특정 아이디 모두의 모든 내용 출력
book_titles = soup2.find_all('p', {"id":"book_title"})
authors = soup2.find_all('p', {"id":"author"})

for book_title, author in zip(book_titles, authors):
    print(book_title.get_text() + '/' + author.get_text())
```

### CSS selector 이용하여 갖고오기  (결과 : 리스트 형태)

```python
print(soup2.select('body h1'))
```

```python
# select는 결과가 리스트 형태이므로 내용만 꺼낼 때 인덱스 명시
print(soup2.select('body h1')[0])
```

```python
print(soup2.select('body p'))
```

```python
print(soup2.select('p'))
```

```python
# 특정 태그의 특정 아이디 불러오기
print(soup2.select('p#book_title'))
```

```python
html3 = """
<!doctype html>
<html>
<head>
<meta charset="utf-8">
<title>사이트 모음</title>
</head>
<body>
<p id="title">
<b>자주 가는 사이트 모음</b>
</p>
<p id="contents">이곳은 자주 가는 사이트를 모아둔 곳입니다.</p>
<a href="http://www.naver.com" class="portal" id="naver">네이버</a> <br>
<a href="https://www.google.com" class="search" id="google">구글</a> <br>
<a href="http://www.daum.net" class="portal" id="daum">다음</a> <br>
<a href="http://www.nl.go.kr" class="government" id="nl">국립중앙도서관</a>
</body>
</html>
"""
```

```python
soup3 = BeautifulSoup(html3, 'lxml')
```

```python
# 특정 태그의 클래스 불러오기
print(soup3.select('a.portal'))
```
